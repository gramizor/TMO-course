Импортируем необходимые библиотеки

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor 
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Admission_Predict_Ver1.1.csv', sep=',')
```

```python
df.head(10)
```

```python
df.info()
```

Разделим выборку на обучающую и тренировочную

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop('Chance of Admit ',axis=1),df['Chance of Admit '],test_size=0.2, random_state=42)
```

Обучим требуемые модели

```python
rgs_lin = DecisionTreeRegressor (random_state=42)
rgs_lin.fit(X_train, y_train)
```

```python
rgs_gbt = GradientBoostingRegressor(n_estimators=200)
rgs_gbt.fit(X_train,y_train)
```

Оценим качество

```python
models = [rgs_lin,rgs_gbt]
for rgs in models:
    print(f"{type(rgs).__name__}:")
    print('\t',f"MSE = {mean_squared_error(y_test,rgs.predict(X_test))}")
    print('\t',f"MAPE = {mean_absolute_percentage_error(y_test,rgs.predict(X_test))}")
```

Сравнивая два метода: дерево решений и градиентный бустинг, видно, что оба показывают хорошие результаты в прогнозировании. Однако, градиентный бустинг продемонстрировал более низкие значения MSE и MAPE по сравнению с деревом решений. Это свидетельствует о том, что градиентный бустинг более точно аппроксимирует целевую переменную и лучше учитывает сложные взаимосвязи между признаками. Таким образом, можно утверждать, что градиентный бустинг эффективнее в данной задаче прогнозирования. При этом стоит отметить, что дерево решений все равно показало приемлемые результаты.
